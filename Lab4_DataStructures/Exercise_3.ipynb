```python
from collections import deque, Counter

CONTROL_NUM = 9
LOCAL_HAZARD = "TSUNAMI"

alerts = deque(maxlen=5)

burst = [
    (CONTROL_NUM, "WEATHER"),
    (CONTROL_NUM + 1, "TRAFFIC"),
    (CONTROL_NUM + 2, "WEATHER"),
    (CONTROL_NUM + 3, "WEATHER"),
    (CONTROL_NUM + 4, "NETWORK"),
    (CONTROL_NUM + 5, "WEATHER"),
    (CONTROL_NUM + 6, "TRAFFIC")
]

extra_alerts = [
    (18, "POWER"),
    (19, "SECURITY"),
    (20, "HEALTH"),
    (21, "RESCUE"),
    (22, "FIRE"),
    (23, "WEATHER"),
    (24, LOCAL_HAZARD),
    (25, "MEDICAL"),
    (26, "FLOOD"),
    (27, "NETWORK"),
    (28, "TRAFFIC")
]

for timestamp, category in burst + extra_alerts:
    current_categories = [c for _, c in alerts]
    category_count = Counter(current_categories)

    if category_count[category] >= 2:
        continue
    else:
        alerts.append((timestamp, category))

weather_at_target = any(ts == CONTROL_NUM + 3 and cat == "WEATHER" for ts, cat in alerts)
local_hazard_present = any(cat == LOCAL_HAZARD for _, cat in alerts)
timestamp_sum = sum(ts for ts, _ in alerts)

print("LOCAL_HAZARD Used:", LOCAL_HAZARD)
print("Total Burst Size (After adding custom alerts):", len(burst + extra_alerts))
print(f'Did "WEATHER" at Timestamp CONTROL_NUM + 3 append? {"Y" if weather_at_target else "N"}')
print(f"Did LOCAL_HAZARD append? {'Y' if local_hazard_present else 'N'}")
print("Final Deque Length:", len(alerts))
print("Final Sum of Timestamps in Deque:", timestamp_sum)
print("Exact Final Deque Output:", list(alerts))
```
